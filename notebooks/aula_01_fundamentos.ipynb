## Exercício 2 — Manipulação direta de pixels

Vamos manipular a matriz da imagem diretamente via NumPy.
Desenhamos um quadrado preto no canto superior esquerdo
e saturamos o canal vermelho no centro da imagem.

```python
# Quadrado preto no canto superior esquerdo
img_quadrado = desenhar_quadrado_preto(img_rgb)
exibir_comparacao(img_rgb, "Original", img_quadrado, "Com quadrado preto")

# Canal vermelho explodido no centro
img_vermelho = explodir_canal_vermelho(img_rgb)
exibir_comparacao(img_rgb, "Original", img_vermelho, "Canal vermelho explodido")
```

## Exercício 3 — Binarização (Thresholding)

Vamos converter a imagem para tons de cinza e aplicar um limiar
manual para transformá-la em preto e branco puro.
Pixels acima de 127 viram branco (255), os demais viram preto (0).

```python
# Binarização com limiar padrão (127)
img_binaria = binarizar(img_rgb)
exibir_comparacao(img_rgb, "Original", img_binaria, "Binarizada (limiar=127)", cmap2="gray")
```

```python
# Comparando diferentes limiares
img_limiar_50 = binarizar(img_rgb, limiar=50)
img_limiar_200 = binarizar(img_rgb, limiar=200)

exibir_comparacao(img_limiar_50, "Limiar 50 (mais branco)", img_limiar_200, "Limiar 200 (mais preto)", cmap1="gray", cmap2="gray")
```
